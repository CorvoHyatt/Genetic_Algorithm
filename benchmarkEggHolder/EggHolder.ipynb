```python
from genetic_algorithm.GA import GeneticAlgorithm as GA
from utils import *
import matplotlib.pyplot as plt
import functools
from math import sin, sqrt
```

# Función Objetivo

```python

precision = 0.001
rango_entero = 512
bits_enteros, bits_decimales = calcular_bits(rango_entero, precision)

def egg_holder(solution,bits_enteros, bits_decimales,precision):
    variables = decodificar_solución(solution,bits_enteros,bits_decimales,precision)
    x , y = variables[0], variables[1]
    return -(y + 47) * sin(sqrt(abs(x/2 + (y + 47)))) - x * sin(sqrt(abs(x - (y + 47))))

fn = functools.partial(egg_holder,bits_enteros = bits_enteros, bits_decimales = bits_decimales, precision=precision)
```

# Experimento 1

```python
parametros = {
    "stopping_criteria_type":"nochangebest",
    "max_nochange_best": 200,
    "population_size": 200,
    "selection_operator": "rank",
    "crossover_operator": "kpoints",
    "mutation_operator":"shuffle",
    "problem_type": "Benchmark",
    "variables": 2,
    "precision":precision,
    "limits":(-512,512),
	"min_or_max": "min",
    "codification": "binary"
}
```

# Prueba 1

```python
ga = GA(**parametros)
```

```python
solve = ga.evolve(fn)
```

```python
print(solve.fitness)
variables = decodificar_solución(solve.genotype,bits_enteros,bits_decimales,precision)
x , y = variables[0], variables[1]
print(f"x: {x} y: {y}")

plt.plot(range(len(ga._cost_worst)), [x for x in ga._cost_worst ], color ='red', label ="Costo Worst")
plt.plot(range(len(ga._cost_best)), [x for x in ga._cost_best ], color ='blue', label ="Costo Best")
plt.plot(range(len(ga._cost_prom)), [x for x in ga._cost_prom ], color ='orange', label ="Costo Promedio")
plt.grid(True, linestyle='dashed')
plt.xlabel('Generaciones')
plt.ylabel('Costo')
plt.title('Genetic Algorithms')
plt.legend()
plt.show()
```

# Prueba 2

```python
ts = TabuSearch(**parametros)
```

```python
solution = ts.fit(fn)
```

```python
variables = decodificar_solución(solution,bits_enteros,bits_decimales,precision)
x , y = variables[0], variables[1]
print(f"x:{x}, y:{y}")
plt.plot(range(len(ts._cost_actual)), [x for x in ts._cost_actual ], color ='orange', label ="Costo Actual")
plt.plot(range(len(ts._cost_best)), [x for x in ts._cost_best ], color ='blue', label ="Costo Best")
plt.grid(True, linestyle='dashed')
plt.xlabel('Iteraciones')
plt.ylabel('Costo')
plt.title('Comparacion del costo actual y el best')
plt.legend()
plt.show()
```
